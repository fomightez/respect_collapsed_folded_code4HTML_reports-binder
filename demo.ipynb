```python
#calculate the area with sympy
import sympy as sp
# Define diameter
d = 150  # mm

# Calculate area using the diameter formula
A = sp.Rational(1, 4) * sp.pi * d**2
A_cm2 = A / 100

#print(f"Area = {A}")  # Exact symbolic form
print(f"Area ≈ {float(A):.2f} mm²")  # Numerical approximation
print(" -or- ")
print(f"Area ≈ {float(A_cm2):.2f} cm²")  # Numerical approximation in centimeters
```

### Relatively cool

$$E = (mc)^2$$

```python
import sympy as sp

# Define symbolic variables
m0, v, c = sp.symbols('m_0 v c', positive=True, real=True)

# Define the Lorentz factor gamma
gamma = 1 / sp.sqrt(1 - (v**2 / c**2))

# Total relativistic energy
E_total = gamma * m0 * c**2

# Compute Taylor expansion up to order v^4 around v = 0
E_expanded = sp.series(E_total, v, 0, n=5).removeO()

print("Expanded Energy Expression:")
sp.pprint(E_expanded)
```

```python
import sympy as sp

m0, v, c, E = sp.symbols('m_0 v c E', positive=True, real=True)

# Define relativistic energy equation
gamma = 1 / sp.sqrt(1 - (v**2 / c**2))
eq = sp.Eq(E, gamma * m0 * c**2)

# Solve for velocity (v)
v_sol = sp.solve(eq, v)

print("Velocity as a function of Total Energy (E):")
sp.pprint(v_sol[0])
```

```python
import sympy as sp

m, c = sp.symbols('m c')
E = m * c**2

# Substitute 1 gram (0.001 kg) and light speed (299,792,458 m/s)
energy_in_joules = E.subs({
    m: 0.001,                 # 1 gram in kg
    c: 299792458              # speed of light in m/s
})

print(f"Energy released from 1g of mass: {float(energy_in_joules):.4e} Joules")
```
